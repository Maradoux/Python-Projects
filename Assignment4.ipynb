```python
#write a program to roll a dice between two users 10 times and the overall winner should be stated 
#Note use any of the loop
#WHILE loop
import random 

user1_score = 0
user2_score = 0
rounds = 1

while rounds <= 10:
    user1_roll = random.randint(1,6)
    user2_roll = random.randint(1,6)
    print(f'Rounds{rounds}:')
    print(f'User1 rolled: {user1_roll}')
    print(f'User2 rolled: {user2_roll}')
    
    if user1_roll > user2_roll:
        print('User1 wins this round!')
        user1_score += 5
    elif user2_roll > user1_roll:
        print('User2 wins this round!')
        user2_score += 5
    else:
        print('This round is a tie!')
        
    rounds +=1
    
if user1_score > user2_score:
    print('User1 is the overall winner!')
elif user2_score > user1_score:
    print('User2 in the overall winner!')
else:
    print('This game is a tie!')
```
