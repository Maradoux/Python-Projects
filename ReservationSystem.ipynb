```python
import string
import random

def get_non_empty_input(prompt):
    while True:
        user_input = input(prompt)
        if user_input.strip(): 
            return user_input   
                
def book():
    while True:
        alength = 4
        numeric_chars = string.digits
        random_string = ''.join(random.choice(numeric_chars) for _ in range(alength))
        print(f'Your Reservation ID is {random_string}')
        RegID = get_non_empty_input('Provide your RegID: ')
        Name = get_non_empty_input('Provide your Full Name: ')
        Email = get_non_empty_input('Provide your Email Address: ')
        Gender = get_non_empty_input('What is your gender(Male/Female/Other)? ')
        PhoneNum = get_non_empty_input('Provide your Phone Number: ')
        Address = get_non_empty_input('Kindly provide you Residential address: ')
        DateOfBooking = get_non_empty_input('Provide Booked date(YYYY/MM/DD): ')
        TimeOfBooking = get_non_empty_input('Provide Booked time(AM/PM): ')
        RoomReserved = get_non_empty_input('What room will you like to book(Standard/Deluxe/Suite): ')

        mycursor.execute('SELECT * FROM records101 WHERE DateOfBooking = %s AND TimeOfBooking = %s', (DateOfBooking, TimeOfBooking))
        result = mycursor.fetchall()
        if result:
            print('A reservation has already been made for the chosen Date and Time. Please choose another.')
        else:
            query = 'INSERT INTO records101(RegID,Name,Email,Gender,PhoneNum,Address,DateOfBooking,TimeOfBooking,RoomReserved) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s)'
            mycursor.execute(query, (RegID, Name, Email, Gender, PhoneNum, Address, DateOfBooking, TimeOfBooking, RoomReserved))
            mydb.commit()
            print('Room Booked successfully...')
            break

def check():
    RegID = int(input('Enter Reservation ID: '))
    query = 'SELECT * FROM records101 WHERE RegID = %s'
    mycursor.execute(query, (RegID,))
    result = mycursor.fetchall()
    if len(result) == 0:
        print('Record not found!')
    else:
        for x in result:
            print(x)

def display_bookings():
    mycursor.execute('SELECT * FROM records101')
    for x in mycursor:
        print(x)

def update_booking():
    RegID = int(input('Enter ReservationID to be updated: '))
    column = input('Enter the column to be updated(Name,Email,Gender,PhoneNum,Address,DateOfBooking,TimeOfBooking,RoomReserved): ')
    newval = input('Enter the new value: ')    
    query = f'UPDATE records101 SET {column} = %s WHERE RegID = %s'
    mycursor.execute(query, (newval, RegID))
    mydb.commit()
    print('Booking updated successfully...')

def delete_booking():
    RegID = int(input('Enter ReservationID to be deleted: '))
    ask = input('Are you sure you want to delete this record(Yes/No)? ')
    if ask.lower() == 'yes':
        query = 'DELETE FROM records101 WHERE RegID = %s'
        mycursor.execute(query, (RegID,))
        mydb.commit()
        print('Booking deleted successfully...')
    else:
        print('Deletion canceled.')

def main():    
    operation = True
    while operation:
        print('Welcome to Prime Hotels!')
        print('Operation[Book, Check, Display, Update, Delete, Logout]')
        option = input('Select Operation: ')
        if option == 'Book'.lower():
            book() 
        elif option == 'Check'.lower():
            check()
        elif option == 'Display'.lower():
            display_bookings()
        elif option == 'Update'.lower():
            update_booking()
        elif option == 'Delete'.lower():
            delete_booking()
        elif option == 'Logout'.lower():
            operation = False
            print('You are Logged out.')
            break
```

```python
main()
```
